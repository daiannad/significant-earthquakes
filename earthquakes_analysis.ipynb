```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# importing database
df = pd.DataFrame()
df = pd.read_csv('database.csv')
df
```

```python
# summary of statistics
df.describe()
```

```python
# calculating the number of missing values
missing_vals = df.isnull().sum()
missing_vals
```

```python
# replacing NaN values with '0'
replace_with = {1:20, 'NaN': '0'}
df.replace(replace_with, inplace = True)
df.fillna(0, inplace = True)
```

```python
# cleaning up string values by removing specified characters
strip_chars = '_!@#$%\^&*()_+'
df = df.applymap(lambda x: x.strip(strip_chars) if isinstance(x, str) else x)
```

```python
df
```

```python
# calculate the number of missing values
missing_vals = df.isnull().sum()
missing_vals
```

```python
# adding a new column 'Category' that categorizes depths into 3 types based on specified bins and labels
min_value = 0
max_value = 700
bins = [min_value, 70, 300, max_value]
labels = ['Shallow', 'Intermediate', 'Deep']
df['Category'] = pd.cut(df['Depth'], bins=bins, labels=labels)
df
```

```python
# reshaping the data 
df_wide = pd.DataFrame({
    "Date": ["2023-01-01", "2023-01-02", "2023-01-03"],
    "Latitude": [34.0522, 40.7128, 51.5074],
    "Longitude": [-118.2437, -74.006, -0.1278],
    "Magnitude": [4.5, 3.8, 5.2],
    "Depth": [10, 8, 12],
    "Category": ["Shallow", "Intermediate", "Deep"]
})
df_wide
```

```python
# reshaping the dataframe with 'melt()' function
df_wide.melt(id_vars=['Date', 'Latitude','Longitude', 'Category'],
             var_name='Magnitude', 
             value_name='Value')
```

```python
# reshaping the dataframe from wide to long format
df_long = pd.melt(df_wide, 
                  id_vars=["Date", "Latitude", "Longitude", "Category"], 
                  value_vars=["Magnitude", "Depth"],
                  var_name="Attribute", 
                  value_name="Value")

print("\nLong format DataFrame:")
print(df_long)
```

```python
# reshaping the data for pivot task
df_wide = pd.DataFrame({
    "Date": ["2023-01-01", "2023-01-02", "2023-01-03"],
    "Latitude": [34.0522, 40.7128, 51.5074],
    "Longitude": [-118.2437, -74.006, -0.1278],
    "Magnitude": [4.5, 3.8, 5.2],
    "Depth": [10, 8, 12],
    "Category": ["Shallow", "Intermediate", "Deep"]
})
df_wide
```

```python
# reshaping the dataframe from wide to long 
df_long = df_wide.melt( 
                  id_vars=["Date", "Latitude", "Longitude", "Category"], 
                  value_vars=["Magnitude", "Depth"],
                  var_name="Measures", 
                  value_name="Value")
df_long
```

```python
# reshaping the dataframe
df_pivot = df_long.pivot_table(
    index=["Date", "Latitude", "Longitude", "Category"],
    columns="Measures",
    values="Value"
).reset_index()
df_pivot
```

```python
# extracting year from the 'Date' column
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
errors = df.loc[df['Date'].isnull(), 'Date']

if not errors.empty:
    print(f"Rows with parsing errors: {errors}")

df['Year'] = df['Date'].dt.year
```

```python
# histogram of the earthquake frequencies by 'Magnitude' column
df['Date'] = pd.to_datetime(df['Date'])
plt.figure(figsize=(10, 6))
plt.hist(df['Magnitude'], bins=30, color='g', edgecolor='black')
plt.title('Earthquake frequencies by magnitude')
plt.xlabel('Magnitude')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# shows a line plot for significant earthquaes per year
earthquakes_per_year = df['Year'].value_counts().sort_index()
fig, ax = plt.subplots(figsize=(15, 10))
ax.plot(earthquakes_per_year.index, earthquakes_per_year.values, marker='o', linestyle='-', color='c', lw=3)
plt.xlabel('Year')
plt.ylabel('Number of Earthquakes')
plt.title('Number of significant earthquakes per year')
plt.xticks(earthquakes_per_year.index, rotation=45)
plt.grid(True)
plt.show()
```

```python
# pie chart that visualizes the distibution of earthquake categories
category_counts = df['Category'].value_counts()
categories = category_counts.index
counts = category_counts.values
plt.figure(figsize=(10, 8))
plt.pie(counts, labels=categories, autopct='%.1f%%', startangle=140)
plt.title('Distribution of earthquake categories')
plt.show()
```

```python
# vartical and horizontal bar plot to visualize the number of earthquakes by categories
category_counts = df['Category'].value_counts()
categories = category_counts.index
counts = category_counts.values
fig, axes = plt.subplots(ncols=2, figsize=plt.figaspect(1/2))
vert_bars = axes[0].bar(categories, counts, color='purple', align='center')
axes[0].set_title('Number of earthquakes by categories')
axes[0].set_ylabel('Number of Earthquakes')
# Horizontal bar plot
horiz_bars = axes[1].barh(categories, counts, color='pink', align='center')
axes[1].set_title('Number of earthquakes by categories')
axes[1].set_xlabel('Number of Earthquakes')
plt.tight_layout()
plt.show()
```
